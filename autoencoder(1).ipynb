```python


import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from sklearn.manifold import TSNE

print("✅ همه کتابخانه‌ها با موفقیت بارگذاری شدند")
print(f"PyTorch version: {torch.__version__}")

# تنظیم device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Device: {device}")

# بارگذاری دیتاست MNIST (برای همه بخش‌ها مشترک است)
transform = transforms.ToTensor()
train_dataset = datasets.MNIST(root='./data', train=True,  download=True, transform=transform)
test_dataset  = datasets.MNIST(root='./data', train=False, download=True, transform=transform)
train_loader  = DataLoader(train_dataset, batch_size=256, shuffle=True)
test_loader   = DataLoader(test_dataset,  batch_size=256, shuffle=False)
print("✅ دیتاست MNIST بارگذاری شد")


# ╔══════════════════════════════════════════════════════════╗
# ║  CELL 1 — خودرمزگذار پایه (Basic Autoencoder)           ║
# ╚══════════════════════════════════════════════════════════╝

print("\n" + "="*55)
print("  بخش ۱: خودرمزگذار پایه")
print("="*55)

class BasicAutoencoder(nn.Module):
    def __init__(self, latent_dim=32):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Flatten(),
            nn.Linear(784, 256),
            nn.ReLU(),
            nn.Linear(256, latent_dim),
            nn.ReLU()
        )
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, 256),
            nn.ReLU(),
            nn.Linear(256, 784),
            nn.Sigmoid()
        )

    def forward(self, x):
        z = self.encoder(x)
        return self.decoder(z), z

def train_model(model, loader, epochs=10, lr=1e-3):
    optimizer = optim.Adam(model.parameters(), lr=lr)
    criterion = nn.MSELoss()
    losses = []
    model.train()
    for epoch in range(epochs):
        epoch_loss = 0
        for imgs, _ in loader:
            imgs = imgs.to(device)
            recon, _ = model(imgs)
            loss = criterion(recon, imgs.view(imgs.size(0), -1))
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            epoch_loss += loss.item()
        avg = epoch_loss / len(loader)
        losses.append(avg)
        print(f"  Epoch {epoch+1:02d}/{epochs} | Loss: {avg:.4f}")
    return losses

basic_ae = BasicAutoencoder(latent_dim=32).to(device)
print("آموزش خودرمزگذار پایه ...")
basic_losses = train_model(basic_ae, train_loader, epochs=10)

# نمودار Loss
plt.figure(figsize=(8, 4))
plt.plot(range(1, len(basic_losses)+1), basic_losses, marker='o', color='steelblue')
plt.title("Basic Autoencoder — Training Loss", fontsize=14)
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# نمایش تصاویر بازسازی‌شده
basic_ae.eval()
with torch.no_grad():
    sample_imgs, _ = next(iter(test_loader))
    sample_imgs = sample_imgs.to(device)
    recon_imgs, _ = basic_ae(sample_imgs)
    recon_imgs = recon_imgs.view(-1, 1, 28, 28)

fig, axes = plt.subplots(2, 10, figsize=(15, 3))
for i in range(10):
    axes[0, i].imshow(sample_imgs[i].cpu().squeeze(), cmap='gray')
    axes[0, i].axis('off')
    axes[1, i].imshow(recon_imgs[i].cpu().squeeze(), cmap='gray')
    axes[1, i].axis('off')
axes[0, 0].set_ylabel("ورودی", fontsize=10)
axes[1, 0].set_ylabel("بازسازی", fontsize=10)
plt.suptitle("Basic Autoencoder — Input vs Reconstruction", fontsize=13)
plt.tight_layout()
plt.show()
print("✅ بخش ۱ تمام شد\n")


# ╔══════════════════════════════════════════════════════════╗
# ║  CELL 2 — خودرمزگذار کاهش‌دهنده نویز (Denoising AE)   ║
# ╚══════════════════════════════════════════════════════════╝

print("="*55)
print("  بخش ۲: خودرمزگذار کاهش‌دهنده نویز")
print("="*55)

class DenoisingAutoencoder(nn.Module):
    def __init__(self, latent_dim=64, noise_factor=0.3):
        super().__init__()
        self.noise_factor = noise_factor
        self.encoder = nn.Sequential(
            nn.Flatten(),
            nn.Linear(784, 400),
            nn.ReLU(),
            nn.Linear(400, latent_dim),
            nn.ReLU()
        )
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, 400),
            nn.ReLU(),
            nn.Linear(400, 784),
            nn.Sigmoid()
        )

    def add_noise(self, x):
        noise = torch.randn_like(x) * self.noise_factor
        return torch.clamp(x + noise, 0., 1.)

    def forward(self, x):
        noisy = self.add_noise(x) if self.training else x
        z = self.encoder(noisy)
        return self.decoder(z), noisy

dae = DenoisingAutoencoder(latent_dim=64, noise_factor=0.4).to(device)
optimizer_dae = optim.Adam(dae.parameters(), lr=1e-3)
criterion = nn.MSELoss()

dae_losses = []
print("آموزش Denoising Autoencoder ...")
for epoch in range(12):
    dae.train()
    epoch_loss = 0
    for imgs, _ in train_loader:
        imgs = imgs.to(device)
        recon, _ = dae(imgs)
        loss = criterion(recon, imgs.view(imgs.size(0), -1))
        optimizer_dae.zero_grad()
        loss.backward()
        optimizer_dae.step()
        epoch_loss += loss.item()
    avg = epoch_loss / len(train_loader)
    dae_losses.append(avg)
    print(f"  Epoch {epoch+1:02d}/12 | Loss: {avg:.4f}")

# نمودار Loss
plt.figure(figsize=(8, 4))
plt.plot(range(1, len(dae_losses)+1), dae_losses, marker='s', color='tomato')
plt.title("Denoising Autoencoder — Training Loss", fontsize=14)
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# نمایش: تصویر اصلی | نویزی | بازسازی‌شده
dae.eval()
with torch.no_grad():
    imgs, _ = next(iter(test_loader))
    imgs = imgs.to(device)
    dae.noise_factor = 0.4
    noisy_imgs = dae.add_noise(imgs)
    recon_dae, _ = dae(noisy_imgs)
    recon_dae = recon_dae.view(-1, 1, 28, 28)

fig, axes = plt.subplots(3, 10, figsize=(15, 4.5))
for i in range(10):
    axes[0, i].imshow(imgs[i].cpu().squeeze(), cmap='gray')
    axes[0, i].axis('off')
    axes[1, i].imshow(noisy_imgs[i].cpu().squeeze(), cmap='gray')
    axes[1, i].axis('off')
    axes[2, i].imshow(recon_dae[i].cpu().squeeze(), cmap='gray')
    axes[2, i].axis('off')
axes[0, 0].set_ylabel("اصلی", fontsize=9)
axes[1, 0].set_ylabel("نویزی", fontsize=9)
axes[2, 0].set_ylabel("بازسازی", fontsize=9)
plt.suptitle("Denoising Autoencoder — Original | Noisy | Reconstructed", fontsize=13)
plt.tight_layout()
plt.show()
print("✅ بخش ۲ تمام شد\n")


# ╔══════════════════════════════════════════════════════════╗
# ║  CELL 3 — خودرمزگذار پراکنده (Sparse Autoencoder)       ║
# ╚══════════════════════════════════════════════════════════╝

print("="*55)
print("  بخش ۳: خودرمزگذار پراکنده")
print("="*55)

class SparseAutoencoder(nn.Module):
    def __init__(self, latent_dim=64, sparsity_target=0.05, beta=3.0):
        super().__init__()
        self.sparsity_target = sparsity_target
        self.beta = beta
        self.encoder = nn.Sequential(
            nn.Flatten(),
            nn.Linear(784, 256),
            nn.ReLU(),
            nn.Linear(256, latent_dim),
            nn.Sigmoid()   # Sigmoid برای محدود کردن به [0,1]
        )
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, 256),
            nn.ReLU(),
            nn.Linear(256, 784),
            nn.Sigmoid()
        )

    def kl_divergence(self, rho_hat):
        rho = self.sparsity_target
        rho_hat = rho_hat.mean(dim=0).clamp(1e-7, 1 - 1e-7)
        return (rho * torch.log(rho / rho_hat) +
                (1 - rho) * torch.log((1 - rho) / (1 - rho_hat))).sum()

    def forward(self, x):
        z = self.encoder(x)
        recon = self.decoder(z)
        return recon, z

sae = SparseAutoencoder(latent_dim=64, sparsity_target=0.05, beta=3.0).to(device)
optimizer_sae = optim.Adam(sae.parameters(), lr=1e-3)
criterion = nn.MSELoss()

sae_losses, sae_recon_losses, sae_sparse_losses = [], [], []
print("آموزش Sparse Autoencoder ...")
for epoch in range(12):
    sae.train()
    ep_total, ep_recon, ep_sparse = 0, 0, 0
    for imgs, _ in train_loader:
        imgs = imgs.to(device)
        recon, z = sae(imgs)
        recon_loss = criterion(recon, imgs.view(imgs.size(0), -1))
        sparse_loss = sae.kl_divergence(z)
        loss = recon_loss + sae.beta * sparse_loss
        optimizer_sae.zero_grad()
        loss.backward()
        optimizer_sae.step()
        ep_total  += loss.item()
        ep_recon  += recon_loss.item()
        ep_sparse += sparse_loss.item()
    n = len(train_loader)
    sae_losses.append(ep_total / n)
    sae_recon_losses.append(ep_recon / n)
    sae_sparse_losses.append(ep_sparse / n)
    print(f"  Epoch {epoch+1:02d}/12 | Total: {ep_total/n:.4f} | Recon: {ep_recon/n:.4f} | Sparse: {ep_sparse/n:.4f}")

# نمودار Loss
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 4))
ax1.plot(sae_losses, marker='o', color='purple', label='Total Loss')
ax1.plot(sae_recon_losses, marker='s', color='steelblue', linestyle='--', label='Recon Loss')
ax1.set_title("Sparse AE — Training Losses", fontsize=13)
ax1.set_xlabel("Epoch"); ax1.set_ylabel("Loss")
ax1.legend(); ax1.grid(True, alpha=0.3)

ax2.plot(sae_sparse_losses, marker='^', color='darkorange', label='Sparsity (KL)')
ax2.set_title("Sparse AE — Sparsity Penalty", fontsize=13)
ax2.set_xlabel("Epoch"); ax2.set_ylabel("KL Loss")
ax2.legend(); ax2.grid(True, alpha=0.3)
plt.tight_layout(); plt.show()

# نمایش فعال‌بودن نورون‌ها
sae.eval()
with torch.no_grad():
    imgs, _ = next(iter(test_loader))
    imgs = imgs.to(device)
    _, z_sample = sae(imgs)
    activation_rate = (z_sample > 0.1).float().mean(dim=0).cpu().numpy()

plt.figure(figsize=(12, 3))
plt.bar(range(len(activation_rate)), activation_rate, color='darkorange', alpha=0.8)
plt.axhline(y=0.05, color='red', linestyle='--', label='Target sparsity (5%)')
plt.title("Sparse AE — Neuron Activation Rate per Latent Dim", fontsize=13)
plt.xlabel("Latent Neuron Index"); plt.ylabel("Activation Rate")
plt.legend(); plt.grid(True, alpha=0.3)
plt.tight_layout(); plt.show()
print("✅ بخش ۳ تمام شد\n")


# ╔══════════════════════════════════════════════════════════╗
# ║  CELL 4 — t-SNE بصری‌سازی فضای نهفته (Basic AE)        ║
# ╚══════════════════════════════════════════════════════════╝

print("="*55)
print("  بخش ۴: بصری‌سازی فضای نهفته با t-SNE (Basic AE)")
print("="*55)

# استخراج latent codes
basic_ae.eval()
latent_codes, labels_list = [], []

with torch.no_grad():
    for imgs, labels in test_loader:
        imgs = imgs.to(device)
        _, z = basic_ae(imgs)
        latent_codes.append(z.cpu().numpy())
        labels_list.append(labels.numpy())

latent_codes = np.concatenate(latent_codes, axis=0)
labels_arr   = np.concatenate(labels_list, axis=0)

# فقط 3000 نمونه برای سرعت
idx = np.random.choice(len(latent_codes), 3000, replace=False)
latent_sample = latent_codes[idx]
labels_sample = labels_arr[idx]

print("در حال اجرای t-SNE (ممکن است چند لحظه طول بکشد) ...")
tsne = TSNE(n_components=2, perplexity=30, n_iter=800, random_state=42)
z_2d = tsne.fit_transform(latent_sample)

colors = plt.cm.get_cmap('tab10', 10)
plt.figure(figsize=(9, 7))
for digit in range(10):
    mask = labels_sample == digit
    plt.scatter(z_2d[mask, 0], z_2d[mask, 1],
                c=[colors(digit)], label=str(digit),
                s=12, alpha=0.7)
plt.title("t-SNE of Basic AE Latent Space (MNIST)", fontsize=14)
plt.xlabel("t-SNE dim 1"); plt.ylabel("t-SNE dim 2")
plt.legend(title="Digit", loc='upper right', markerscale=2)
plt.grid(True, alpha=0.2)
plt.tight_layout()
plt.show()
print("✅ بخش ۴ تمام شد\n")


# ╔══════════════════════════════════════════════════════════╗
# ║  CELL 5 — خودرمزگذار واریانسی (VAE)                    ║
# ╚══════════════════════════════════════════════════════════╝

print("="*55)
print("  بخش ۵: خودرمزگذار واریانسی (VAE)")
print("="*55)

class VAE(nn.Module):
    def __init__(self, latent_dim=16):
        super().__init__()
        self.latent_dim = latent_dim
        # Encoder
        self.enc_fc1 = nn.Linear(784, 400)
        self.enc_mu  = nn.Linear(400, latent_dim)
        self.enc_logvar = nn.Linear(400, latent_dim)
        # Decoder
        self.dec_fc1 = nn.Linear(latent_dim, 400)
        self.dec_fc2 = nn.Linear(400, 784)

    def encode(self, x):
        h = torch.relu(self.enc_fc1(x.view(-1, 784)))
        return self.enc_mu(h), self.enc_logvar(h)

    def reparameterize(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std

    def decode(self, z):
        h = torch.relu(self.dec_fc1(z))
        return torch.sigmoid(self.dec_fc2(h))

    def forward(self, x):
        mu, logvar = self.encode(x)
        z = self.reparameterize(mu, logvar)
        return self.decode(z), mu, logvar

def vae_loss(recon, x, mu, logvar, beta=1.0):
    recon_loss = nn.functional.binary_cross_entropy(
        recon, x.view(-1, 784), reduction='sum') / x.size(0)
    kl_loss = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp()) / x.size(0)
    return recon_loss + beta * kl_loss, recon_loss, kl_loss

vae = VAE(latent_dim=16).to(device)
optimizer_vae = optim.Adam(vae.parameters(), lr=1e-3)

vae_total_losses, vae_recon_losses, vae_kl_losses = [], [], []
print("آموزش VAE ...")
for epoch in range(15):
    vae.train()
    ep_total, ep_recon, ep_kl = 0, 0, 0
    for imgs, _ in train_loader:
        imgs = imgs.to(device)
        recon, mu, logvar = vae(imgs)
        loss, rl, kl = vae_loss(recon, imgs, mu, logvar, beta=1.0)
        optimizer_vae.zero_grad()
        loss.backward()
        optimizer_vae.step()
        ep_total += loss.item()
        ep_recon += rl.item()
        ep_kl    += kl.item()
    n = len(train_loader)
    vae_total_losses.append(ep_total / n)
    vae_recon_losses.append(ep_recon / n)
    vae_kl_losses.append(ep_kl / n)
    print(f"  Epoch {epoch+1:02d}/15 | Total: {ep_total/n:.1f} | Recon: {ep_recon/n:.1f} | KL: {ep_kl/n:.3f}")

# نمودار ELBO
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 4))
ax1.plot(vae_total_losses, marker='o', color='darkgreen', label='ELBO (Total)')
ax1.plot(vae_recon_losses, marker='s', color='steelblue', linestyle='--', label='Reconstruction')
ax1.set_title("VAE — Training Loss (ELBO)", fontsize=13)
ax1.set_xlabel("Epoch"); ax1.set_ylabel("Loss")
ax1.legend(); ax1.grid(True, alpha=0.3)

ax2.plot(vae_kl_losses, marker='^', color='crimson', label='KL Divergence')
ax2.set_title("VAE — KL Divergence", fontsize=13)
ax2.set_xlabel("Epoch"); ax2.set_ylabel("KL Loss")
ax2.legend(); ax2.grid(True, alpha=0.3)
plt.tight_layout(); plt.show()

# تولید تصاویر جدید (نمونه از N(0,1))
vae.eval()
with torch.no_grad():
    z_random = torch.randn(20, vae.latent_dim).to(device)
    generated = vae.decode(z_random).view(-1, 1, 28, 28).cpu()

fig, axes = plt.subplots(2, 10, figsize=(15, 3))
for i in range(20):
    axes[i // 10, i % 10].imshow(generated[i].squeeze(), cmap='gray')
    axes[i // 10, i % 10].axis('off')
plt.suptitle("VAE — Generated Samples from N(0,1)", fontsize=13)
plt.tight_layout(); plt.show()
print("✅ بخش ۵ تمام شد\n")


# ╔══════════════════════════════════════════════════════════╗
# ║  CELL 6 — t-SNE بصری‌سازی فضای نهفته VAE + Interpolation ║
# ╚══════════════════════════════════════════════════════════╝

print("="*55)
print("  بخش ۶: t-SNE فضای نهفته VAE + درون‌یابی")
print("="*55)

# استخراج mu از VAE
vae.eval()
vae_latents, vae_labels = [], []
with torch.no_grad():
    for imgs, labels in test_loader:
        imgs = imgs.to(device)
        mu, _ = vae.encode(imgs)
        vae_latents.append(mu.cpu().numpy())
        vae_labels.append(labels.numpy())

vae_latents = np.concatenate(vae_latents, axis=0)
vae_labels  = np.concatenate(vae_labels, axis=0)

idx2 = np.random.choice(len(vae_latents), 3000, replace=False)
print("در حال اجرای t-SNE برای VAE ...")
tsne2 = TSNE(n_components=2, perplexity=30, n_iter=800, random_state=0)
z_2d_vae = tsne2.fit_transform(vae_latents[idx2])

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
colors = plt.cm.get_cmap('tab10', 10)
for digit in range(10):
    mask = vae_labels[idx2] == digit
    ax1.scatter(z_2d_vae[mask, 0], z_2d_vae[mask, 1],
                c=[colors(digit)], label=str(digit), s=12, alpha=0.7)
ax1.set_title("t-SNE of VAE Latent Space (μ)", fontsize=13)
ax1.set_xlabel("t-SNE dim 1"); ax1.set_ylabel("t-SNE dim 2")
ax1.legend(title="Digit", markerscale=2)
ax1.grid(True, alpha=0.2)

# درون‌یابی بین دو رقم
with torch.no_grad():
    # پیدا کردن یک نمونه از ۱ و یک از ۷
    test_imgs, test_lbls = next(iter(test_loader))
    idx_1 = (test_lbls == 1).nonzero(as_tuple=True)[0][0]
    idx_7 = (test_lbls == 7).nonzero(as_tuple=True)[0][0]

    img1 = test_imgs[idx_1].unsqueeze(0).to(device)
    img7 = test_imgs[idx_7].unsqueeze(0).to(device)

    mu1, _ = vae.encode(img1)
    mu7, _ = vae.encode(img7)

    n_steps = 10
    interp_imgs = []
    for alpha in np.linspace(0, 1, n_steps):
        z_interp = (1 - alpha) * mu1 + alpha * mu7
        img_interp = vae.decode(z_interp).view(1, 28, 28).cpu()
        interp_imgs.append(img_interp)

for i, img in enumerate(interp_imgs):
    ax2.add_patch(plt.Rectangle((i * 2.9, 0), 2.8, 2.8, fill=False))
    ax2_inner = fig.add_axes([0.52 + i * 0.046, 0.15, 0.04, 0.7])
    ax2_inner.imshow(img.squeeze(), cmap='gray')
    ax2_inner.axis('off')

ax2.set_xlim(0, 10); ax2.set_ylim(0, 1)
ax2.axis('off')
ax2.set_title("VAE — Interpolation: 1 → 7", fontsize=13, y=1.02)
plt.tight_layout(); plt.show()
print("✅ بخش ۶ تمام شد\n")


# ╔══════════════════════════════════════════════════════════╗
# ║  CELL 7 — مقایسه نهایی همه مدل‌ها                      ║
# ╚══════════════════════════════════════════════════════════╝

print("="*55)
print("  بخش ۷: مقایسه نهایی همه مدل‌ها")
print("="*55)

fig, axes = plt.subplots(2, 4, figsize=(16, 7))

# --- Basic AE ---
basic_ae.eval()
with torch.no_grad():
    imgs_cmp, _ = next(iter(test_loader))
    imgs_cmp = imgs_cmp.to(device)
    recon_b, _ = basic_ae(imgs_cmp)
    recon_b = recon_b.view(-1, 1, 28, 28)

for i in range(4):
    axes[0, i].imshow(imgs_cmp[i].cpu().squeeze(), cmap='gray')
    axes[0, i].set_title(f"Original {i+1}", fontsize=9)
    axes[0, i].axis('off')
    axes[1, i].imshow(recon_b[i].cpu().squeeze(), cmap='gray')
    axes[1, i].set_title(f"Basic AE Recon {i+1}", fontsize=9)
    axes[1, i].axis('off')

plt.suptitle("مقایسه: تصاویر اصلی در مقابل بازسازی Basic AE", fontsize=13, y=1.01)
plt.tight_layout(); plt.show()

# جدول مقایسه Loss
print("\n" + "="*50)
print("  📊 خلاصه نتایج آموزش")
print("="*50)
print(f"  Basic AE    — Final Loss : {basic_losses[-1]:.4f}")
print(f"  Denoising AE— Final Loss : {dae_losses[-1]:.4f}")
print(f"  Sparse AE   — Final Loss : {sae_losses[-1]:.4f}")
print(f"  VAE         — Final ELBO : {vae_total_losses[-1]:.1f}  (KL: {vae_kl_losses[-1]:.3f})")
print("="*50)

# نمودار مقایسه Loss همه مدل‌ها (نرمال‌شده)
plt.figure(figsize=(10, 5))
n1 = np.array(basic_losses) / basic_losses[0]
n2 = np.array(dae_losses) / dae_losses[0]
n3 = np.array(sae_losses) / sae_losses[0]
n4 = np.array(vae_total_losses) / vae_total_losses[0]
max_ep = max(len(n1), len(n2), len(n3), len(n4))
plt.plot(range(1, len(n1)+1), n1, marker='o', label='Basic AE',       color='steelblue')
plt.plot(range(1, len(n2)+1), n2, marker='s', label='Denoising AE',   color='tomato')
plt.plot(range(1, len(n3)+1), n3, marker='^', label='Sparse AE',      color='purple')
plt.plot(range(1, len(n4)+1), n4, marker='D', label='VAE (ELBO)',     color='darkgreen')
plt.title("مقایسه روند کاهش Loss (نرمال‌شده)", fontsize=14)
plt.xlabel("Epoch"); plt.ylabel("Normalized Loss")
plt.legend(); plt.grid(True, alpha=0.3)
plt.tight_layout(); plt.show()

print("\n🎉 همه بخش‌ها با موفقیت اجرا شدند!")
```
