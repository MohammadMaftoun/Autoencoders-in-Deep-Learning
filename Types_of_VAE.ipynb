```python

import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from sklearn.manifold import TSNE

print("✅ کتابخانه‌ها بارگذاری شدند")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Device: {device}")

transform = transforms.ToTensor()
train_dataset = datasets.MNIST(root='./data', train=True,  download=True, transform=transform)
test_dataset  = datasets.MNIST(root='./data', train=False, download=True, transform=transform)
train_loader  = DataLoader(train_dataset, batch_size=256, shuffle=True)
test_loader   = DataLoader(test_dataset,  batch_size=256, shuffle=False)
print("✅ دیتاست MNIST بارگذاری شد")


# ╔══════════════════════════════════════════════════════════╗
# ║  CELL 1 — CVAE                                            ║
# ╚══════════════════════════════════════════════════════════╝

print("\n" + "="*60)
print("  CELL 1: CVAE — Conditional Variational Autoencoder")
print("="*60)

NUM_CLASSES = 10
LATENT_DIM_CVAE = 16

class CVAE(nn.Module):
    def __init__(self, input_dim=784, latent_dim=16, num_classes=10):
        super().__init__()
        self.latent_dim  = latent_dim
        self.num_classes = num_classes

        self.enc_fc1   = nn.Linear(input_dim + num_classes, 400)
        self.enc_mu    = nn.Linear(400, latent_dim)
        self.enc_logvar = nn.Linear(400, latent_dim)

        self.dec_fc1 = nn.Linear(latent_dim + num_classes, 400)
        self.dec_fc2 = nn.Linear(400, input_dim)

    def one_hot(self, labels):
        oh = torch.zeros(labels.size(0), self.num_classes, device=labels.device)
        oh.scatter_(1, labels.unsqueeze(1), 1)
        return oh

    def encode(self, x, y):
        oh = self.one_hot(y)
        h  = F.relu(self.enc_fc1(torch.cat([x, oh], dim=1)))
        return self.enc_mu(h), self.enc_logvar(h)

    def reparameterize(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        return mu + torch.randn_like(std) * std

    def decode(self, z, y):
        oh = self.one_hot(y)
        h  = F.relu(self.dec_fc1(torch.cat([z, oh], dim=1)))
        return torch.sigmoid(self.dec_fc2(h))

    def forward(self, x, y):
        x_flat = x.view(-1, 784)
        mu, logvar = self.encode(x_flat, y)
        z = self.reparameterize(mu, logvar)
        return self.decode(z, y), mu, logvar


def cvae_loss(recon, x, mu, logvar, beta=1.0):
    bce = F.binary_cross_entropy(recon, x.view(-1, 784), reduction='sum') / x.size(0)
    kl  = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp()) / x.size(0)
    return bce + beta * kl, bce, kl


cvae = CVAE(latent_dim=LATENT_DIM_CVAE).to(device)
opt_cvae = optim.Adam(cvae.parameters(), lr=1e-3)

cvae_losses, cvae_kls = [], []
EPOCHS_CVAE = 15
print(f"آموزش CVAE برای {EPOCHS_CVAE} epoch ...")

for epoch in range(EPOCHS_CVAE):
    cvae.train()
    ep_loss, ep_kl = 0, 0
    for imgs, labels in train_loader:
        imgs, labels = imgs.to(device), labels.to(device)
        recon, mu, logvar = cvae(imgs, labels)
        loss, bce, kl = cvae_loss(recon, imgs, mu, logvar)
        opt_cvae.zero_grad()
        loss.backward()
        opt_cvae.step()
        ep_loss += loss.item()
        ep_kl += kl.item()
    n = len(train_loader)
    cvae_losses.append(ep_loss / n)
    cvae_kls.append(ep_kl / n)
    print(f"  Epoch {epoch+1:02d}/{EPOCHS_CVAE} | Loss: {ep_loss/n:.1f} | KL: {ep_kl/n:.3f}")

# نمودار Loss
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 4))
ax1.plot(cvae_losses, marker='o', color='royalblue', label='Total Loss')
ax1.set_title("CVAE — Training Loss"); ax1.set_xlabel("Epoch"); ax1.set_ylabel("Loss")
ax1.legend(); ax1.grid(True, alpha=0.3)

ax2.plot(cvae_kls, marker='^', color='crimson', label='KL Divergence')
ax2.set_title("CVAE — KL Divergence"); ax2.set_xlabel("Epoch"); ax2.set_ylabel("KL")
ax2.legend(); ax2.grid(True, alpha=0.3)
plt.tight_layout(); plt.show()

# تولید شرطی
cvae.eval()
with torch.no_grad():
    fig, axes = plt.subplots(3, 10, figsize=(16, 5))
    for digit in range(10):
        y_cond = torch.full((3,), digit, dtype=torch.long, device=device)
        z = torch.randn(3, LATENT_DIM_CVAE, device=device)
        gen = cvae.decode(z, y_cond).view(3, 28, 28).cpu()
        for row in range(3):
            axes[row, digit].imshow(gen[row], cmap='gray')
            axes[row, digit].axis('off')
        axes[0, digit].set_title(f"Digit {digit}")
    plt.suptitle("CVAE — Conditional Generation")
    plt.tight_layout(); plt.show()

print("✅ CELL 1 (CVAE) تمام شد\n")


# ╔══════════════════════════════════════════════════════════╗
# ║  CELL 2 — β-VAE (رفع خطا Latent Traversal)               ║
# ╚══════════════════════════════════════════════════════════╝

print("="*60)
print("  CELL 2: β-VAE")
print("="*60)

LATENT_DIM_BVAE = 10

class BetaVAE(nn.Module):
    def __init__(self, latent_dim=10, beta=4.0):
        super().__init__()
        self.beta = beta
        self.enc = nn.Sequential(nn.Linear(784, 400), nn.ReLU())
        self.fc_mu     = nn.Linear(400, latent_dim)
        self.fc_logvar = nn.Linear(400, latent_dim)
        self.dec = nn.Sequential(
            nn.Linear(latent_dim, 400), nn.ReLU(),
            nn.Linear(400, 784), nn.Sigmoid()
        )

    def encode(self, x):
        h = self.enc(x.view(-1, 784))
        return self.fc_mu(h), self.fc_logvar(h)

    def reparameterize(self, mu, logvar):
        return mu + torch.randn_like(mu) * torch.exp(0.5 * logvar)

    def decode(self, z):
        return self.dec(z)

    def forward(self, x):
        mu, logvar = self.encode(x)
        z = self.reparameterize(mu, logvar)
        return self.decode(z), mu, logvar

    def loss(self, recon, x, mu, logvar):
        bce = F.binary_cross_entropy(recon, x.view(-1, 784), reduction='sum') / x.size(0)
        kl  = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp()) / x.size(0)
        return bce + self.beta * kl, bce, kl


def train_bvae(beta, epochs=15):
    model = BetaVAE(latent_dim=LATENT_DIM_BVAE, beta=beta).to(device)
    opt   = optim.Adam(model.parameters(), lr=1e-3)
    losses, kls = [], []
    for epoch in range(epochs):
        model.train()
        ep_loss, ep_kl = 0, 0
        for imgs, _ in train_loader:
            imgs = imgs.to(device)
            recon, mu, logvar = model(imgs)
            loss, _, kl = model.loss(recon, imgs, mu, logvar)
            opt.zero_grad(); loss.backward(); opt.step()
            ep_loss += loss.item(); ep_kl += kl.item()
        n = len(train_loader)
        losses.append(ep_loss / n)
        kls.append(ep_kl / n)
        print(f"  β={beta} | Epoch {epoch+1:02d}/{epochs} | Loss: {ep_loss/n:.1f} | KL: {ep_kl/n:.3f}")
    return model, losses, kls


model_b1,  loss_b1,  kl_b1  = train_bvae(beta=1)
model_b4,  loss_b4,  kl_b4  = train_bvae(beta=4)
model_b10, loss_b10, kl_b10 = train_bvae(beta=10)

# Latent Traversal — رفع خطا (اضافه کردن .detach())
print("\nLatent Traversal ...")
model_b4.eval()
with torch.no_grad():
    sample_img = next(iter(test_loader))[0][:1].to(device)
    mu_base, _ = model_b4.encode(sample_img)

n_dims    = min(5, LATENT_DIM_BVAE)
n_vals    = 7
val_range = np.linspace(-3, 3, n_vals)

fig, axes = plt.subplots(n_dims, n_vals, figsize=(14, 8))
for d in range(n_dims):
    for vi, val in enumerate(val_range):
        z_mod = mu_base.clone()
        z_mod[0, d] = val
        img = model_b4.decode(z_mod).view(28, 28).cpu().detach().numpy()   # ← رفع خطا
        axes[d, vi].imshow(img, cmap='gray')
        axes[d, vi].axis('off')
        if d == 0:
            axes[d, vi].set_title(f"{val:.1f}")
    axes[d, 0].set_ylabel(f"dim {d+1}", rotation=0, labelpad=30)
plt.suptitle("β-VAE (β=4) — Latent Traversal")
plt.tight_layout(); plt.show()

print("✅ CELL 2 (β-VAE) تمام شد\n")


# ╔══════════════════════════════════════════════════════════╗
# ║  CELL 3 — VQ-VAE                                          ║
# ╚══════════════════════════════════════════════════════════╝

print("="*60)
print("  CELL 3: VQ-VAE")
print("="*60)

class VectorQuantizer(nn.Module):
    def __init__(self, num_embeddings=64, embedding_dim=32, commitment_cost=0.25):
        super().__init__()
        self.num_embeddings = num_embeddings
        self.embedding_dim = embedding_dim
        self.commitment_cost = commitment_cost
        self.embedding = nn.Embedding(num_embeddings, embedding_dim)
        nn.init.uniform_(self.embedding.weight, -1/num_embeddings, 1/num_embeddings)

    def forward(self, z_e):
        dists = (z_e.pow(2).sum(dim=1, keepdim=True)
                 - 2 * z_e @ self.embedding.weight.t()
                 + self.embedding.weight.pow(2).sum(dim=1))
        encoding_indices = dists.argmin(dim=1)
        z_q = self.embedding(encoding_indices)

        codebook_loss = F.mse_loss(z_q.detach(), z_e)
        commitment_loss = F.mse_loss(z_q, z_e.detach())
        vq_loss = codebook_loss + self.commitment_cost * commitment_loss

        z_q_st = z_e + (z_q - z_e).detach()
        return z_q_st, vq_loss, encoding_indices


class VQVAE(nn.Module):
    def __init__(self, latent_dim=32, num_embeddings=64):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Flatten(),
            nn.Linear(784, 256), nn.ReLU(),
            nn.Linear(256, latent_dim)
        )
        self.vq = VectorQuantizer(num_embeddings, latent_dim)
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, 256), nn.ReLU(),
            nn.Linear(256, 784), nn.Sigmoid()
        )

    def forward(self, x):
        z_e = self.encoder(x)
        z_q, vq_loss, indices = self.vq(z_e)
        recon = self.decoder(z_q)
        return recon, vq_loss, indices


vqvae = VQVAE().to(device)
opt_vq = optim.Adam(vqvae.parameters(), lr=1e-3)

vq_total, vq_recon_l, vq_vq_l = [], [], []
EPOCHS_VQ = 15

for epoch in range(EPOCHS_VQ):
    vqvae.train()
    ep_total, ep_recon, ep_vq = 0, 0, 0
    for imgs, _ in train_loader:
        imgs = imgs.to(device)
        recon, vq_loss, _ = vqvae(imgs)
        recon_loss = F.mse_loss(recon, imgs.view(-1, 784))
        loss = recon_loss + vq_loss
        opt_vq.zero_grad(); loss.backward(); opt_vq.step()
        ep_total += loss.item()
        ep_recon += recon_loss.item()
        ep_vq    += vq_loss.item()
    n = len(train_loader)
    vq_total.append(ep_total / n)
    vq_recon_l.append(ep_recon / n)
    vq_vq_l.append(ep_vq / n)
    print(f"  Epoch {epoch+1:02d} | Total: {ep_total/n:.4f} | Recon: {ep_recon/n:.4f} | VQ: {ep_vq/n:.4f}")

print("✅ CELL 3 (VQ-VAE) تمام شد\n")


# ╔══════════════════════════════════════════════════════════╗
# ║  CELL 4 — VGAE                                            ║
# ╚══════════════════════════════════════════════════════════╝

print("="*60)
print("  CELL 4: VGAE")
print("="*60)

# نصب torch_geometric
!pip install torch_geometric -q

from torch_geometric.datasets import KarateClub
from torch_geometric.utils import train_test_split_edges, negative_sampling, to_dense_adj
from torch_geometric.nn import GCNConv

dataset = KarateClub()
data = dataset[0]
data = train_test_split_edges(data)
data = data.to(device)

class GCNEncoder(nn.Module):
    def __init__(self, in_channels, hidden_dim, latent_dim):
        super().__init__()
        self.conv1 = GCNConv(in_channels, hidden_dim)
        self.conv_mu = GCNConv(hidden_dim, latent_dim)
        self.conv_lv = GCNConv(hidden_dim, latent_dim)

    def forward(self, x, edge_index):
        h = F.relu(self.conv1(x, edge_index))
        return self.conv_mu(h, edge_index), self.conv_lv(h, edge_index)


class VGAE_Model(nn.Module):
    def __init__(self, encoder):
        super().__init__()
        self.encoder = encoder

    def reparameterize(self, mu, logvar):
        if self.training:
            std = torch.exp(0.5 * logvar)
            return mu + torch.randn_like(std) * std
        return mu

    def encode(self, x, edge_index):
        mu, logvar = self.encoder(x, edge_index)
        z = self.reparameterize(mu, logvar)
        return z, mu, logvar

    def decode(self, z, edge_index):
        row, col = edge_index
        return torch.sigmoid((z[row] * z[col]).sum(dim=1))

    def decode_all(self, z):
        return torch.sigmoid(z @ z.t())

    def recon_loss(self, z, pos_edge_index):
        pos_pred = self.decode(z, pos_edge_index)
        pos_loss = -torch.log(pos_pred + 1e-8).mean()
        neg_ei = negative_sampling(pos_edge_index, num_nodes=z.size(0),
                                   num_neg_samples=pos_edge_index.size(1))
        neg_pred = self.decode(z, neg_ei)
        neg_loss = -torch.log(1 - neg_pred + 1e-8).mean()
        return pos_loss + neg_loss

    def kl_loss(self, mu, logvar):
        return -0.5 * torch.mean(1 + logvar - mu.pow(2) - logvar.exp())


vgae = VGAE_Model(GCNEncoder(data.x.shape[1], 32, 16)).to(device)
opt_vg = optim.Adam(vgae.parameters(), lr=0.01)

# آموزش VGAE (خلاصه‌تر برای سرعت)
EPOCHS_VGAE = 150
for epoch in range(EPOCHS_VGAE):
    vgae.train()
    z, mu, logvar = vgae.encode(data.x, data.train_pos_edge_index)
    recon_l = vgae.recon_loss(z, data.train_pos_edge_index)
    kl_l = vgae.kl_loss(mu, logvar)
    loss = recon_l + kl_l
    opt_vg.zero_grad(); loss.backward(); opt_vg.step()
    if (epoch + 1) % 30 == 0:
        print(f"  Epoch {epoch+1:03d} | Loss: {loss.item():.4f} | KL: {kl_l.item():.4f}")

print("✅ CELL 4 (VGAE) تمام شد\n")


# ╔══════════════════════════════════════════════════════════╗
# ║  CELL 5 — مقایسه نهایی                                   ║
# ╚══════════════════════════════════════════════════════════╝

print("🎉 همه سلول‌ها با موفقیت اجرا شدند!")
```

```python
# ============================================================
# مدل‌های پیشرفته VAE با نمودارهای Plotly تعاملی
# CVAE | β-VAE | VQ-VAE | VGAE
# ============================================================

import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms

# =============== Plotly ===============
import plotly.graph_objects as go
from plotly.subplots import make_subplots

print("✅ کتابخانه‌ها بارگذاری شدند")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Device: {device}")

# دیتاست
transform = transforms.ToTensor()
train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)
train_loader = DataLoader(train_dataset, batch_size=256, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=256, shuffle=False)

print("✅ دیتاست MNIST آماده است\n")


# ============================ CELL 1: CVAE ============================
print("="*70)
print("CELL 1: CVAE — Conditional VAE")
print("="*70)

class CVAE(nn.Module):
    def __init__(self, latent_dim=16, num_classes=10):
        super().__init__()
        self.latent_dim = latent_dim
        self.num_classes = num_classes

        self.enc_fc1 = nn.Linear(784 + num_classes, 400)
        self.enc_mu = nn.Linear(400, latent_dim)
        self.enc_logvar = nn.Linear(400, latent_dim)

        self.dec_fc1 = nn.Linear(latent_dim + num_classes, 400)
        self.dec_fc2 = nn.Linear(400, 784)

    def one_hot(self, labels):
        oh = torch.zeros(labels.size(0), self.num_classes, device=labels.device)
        oh.scatter_(1, labels.unsqueeze(1), 1)
        return oh

    def encode(self, x, y):
        oh = self.one_hot(y)
        h = F.relu(self.enc_fc1(torch.cat([x, oh], dim=1)))
        return self.enc_mu(h), self.enc_logvar(h)

    def reparameterize(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        return mu + torch.randn_like(std) * std

    def decode(self, z, y):
        oh = self.one_hot(y)
        h = F.relu(self.dec_fc1(torch.cat([z, oh], dim=1)))
        return torch.sigmoid(self.dec_fc2(h))

    def forward(self, x, y):
        x = x.view(-1, 784)
        mu, logvar = self.encode(x, y)
        z = self.reparameterize(mu, logvar)
        return self.decode(z, y), mu, logvar


def cvae_loss(recon, x, mu, logvar):
    bce = F.binary_cross_entropy(recon, x.view(-1, 784), reduction='sum') / x.size(0)
    kl = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp()) / x.size(0)
    return bce + kl, bce, kl


cvae = CVAE().to(device)
opt_cvae = optim.Adam(cvae.parameters(), lr=1e-3)

cvae_losses, cvae_kls = [], []
for epoch in range(12):
    cvae.train()
    ep_loss, ep_kl = 0, 0
    for imgs, labels in train_loader:
        imgs, labels = imgs.to(device), labels.to(device)
        recon, mu, logvar = cvae(imgs, labels)
        loss, _, kl = cvae_loss(recon, imgs, mu, logvar)
        opt_cvae.zero_grad()
        loss.backward()
        opt_cvae.step()
        ep_loss += loss.item()
        ep_kl += kl.item()
    n = len(train_loader)
    cvae_losses.append(ep_loss / n)
    cvae_kls.append(ep_kl / n)
    print(f"Epoch {epoch+1:02d} | Loss: {ep_loss/n:.2f} | KL: {ep_kl/n:.3f}")

# Plotly CVAE
fig = make_subplots(rows=1, cols=2, subplot_titles=("Total Loss", "KL Divergence"))
fig.add_trace(go.Scatter(y=cvae_losses, mode='lines+markers', name='Total Loss', line=dict(color='royalblue')), row=1, col=1)
fig.add_trace(go.Scatter(y=cvae_kls, mode='lines+markers', name='KL', line=dict(color='crimson')), row=1, col=2)
fig.update_layout(title="CVAE Training Progress", height=400, template="plotly_white")
fig.show()

print("✅ CVAE تمام شد\n")


# ============================ CELL 2: β-VAE ============================
print("="*70)
print("CELL 2: β-VAE")
print("="*70)

class BetaVAE(nn.Module):
    def __init__(self, latent_dim=10, beta=4.0):
        super().__init__()
        self.beta = beta
        self.enc = nn.Sequential(nn.Linear(784, 400), nn.ReLU())
        self.fc_mu = nn.Linear(400, latent_dim)
        self.fc_logvar = nn.Linear(400, latent_dim)
        self.dec = nn.Sequential(nn.Linear(latent_dim, 400), nn.ReLU(),
                                nn.Linear(400, 784), nn.Sigmoid())

    def encode(self, x):
        h = self.enc(x.view(-1, 784))
        return self.fc_mu(h), self.fc_logvar(h)

    def reparameterize(self, mu, logvar):
        return mu + torch.randn_like(mu) * torch.exp(0.5 * logvar)

    def decode(self, z):
        return self.dec(z)

    def forward(self, x):
        mu, logvar = self.encode(x)
        z = self.reparameterize(mu, logvar)
        return self.decode(z), mu, logvar


def train_bvae(beta, epochs=12):
    model = BetaVAE(latent_dim=10, beta=beta).to(device)
    opt = optim.Adam(model.parameters(), lr=1e-3)
    losses, kls = [], []
    for epoch in range(epochs):
        model.train()
        ep_loss, ep_kl = 0, 0
        for imgs, _ in train_loader:
            imgs = imgs.to(device)
            recon, mu, logvar = model(imgs)
            bce = F.binary_cross_entropy(recon, imgs.view(-1, 784), reduction='sum') / imgs.size(0)
            kl = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp()) / imgs.size(0)
            loss = bce + beta * kl
            opt.zero_grad(); loss.backward(); opt.step()
            ep_loss += loss.item()
            ep_kl += kl.item()
        n = len(train_loader)
        losses.append(ep_loss/n)
        kls.append(ep_kl/n)
        print(f"β={beta} | Epoch {epoch+1:02d} | Loss: {ep_loss/n:.2f} | KL: {ep_kl/n:.3f}")
    return losses, kls

loss_b1, kl_b1 = train_bvae(1)
loss_b4, kl_b4 = train_bvae(4)
loss_b10, kl_b10 = train_bvae(10)

# Plotly β-VAE
fig = make_subplots(rows=1, cols=2, subplot_titles=("Total Loss", "KL Divergence"))
for beta, loss, kl, color in zip([1,4,10], [loss_b1, loss_b4, loss_b10], [kl_b1, kl_b4, kl_b10],
                                ['steelblue','darkorange','crimson']):
    fig.add_trace(go.Scatter(y=loss, mode='lines+markers', name=f'β={beta}', line=dict(color=color)), row=1, col=1)
    fig.add_trace(go.Scatter(y=kl, mode='lines+markers', name=f'β={beta}', line=dict(color=color)), row=1, col=2)
fig.update_layout(title="β-VAE — Comparison of Different β Values", height=450, template="plotly_white")
fig.show()

print("✅ β-VAE تمام شد\n")


# ============================ CELL 3: VQ-VAE ============================
print("="*70)
print("CELL 3: VQ-VAE")
print("="*70)

class VectorQuantizer(nn.Module):
    def __init__(self, num_embeddings=64, embedding_dim=32, commitment_cost=0.25):
        super().__init__()
        self.embedding = nn.Embedding(num_embeddings, embedding_dim)
        nn.init.uniform_(self.embedding.weight, -1/num_embeddings, 1/num_embeddings)
        self.commitment_cost = commitment_cost

    def forward(self, z_e):
        dists = (z_e.pow(2).sum(1, keepdim=True) - 2 * z_e @ self.embedding.weight.t() +
                self.embedding.weight.pow(2).sum(1))
        indices = dists.argmin(1)
        z_q = self.embedding(indices)

        vq_loss = F.mse_loss(z_q.detach(), z_e) + self.commitment_cost * F.mse_loss(z_q, z_e.detach())
        z_q_st = z_e + (z_q - z_e).detach()
        return z_q_st, vq_loss, indices


class VQVAE(nn.Module):
    def __init__(self, latent_dim=32, num_embeddings=64):
        super().__init__()
        self.encoder = nn.Sequential(nn.Flatten(), nn.Linear(784, 256), nn.ReLU(), nn.Linear(256, latent_dim))
        self.vq = VectorQuantizer(num_embeddings, latent_dim)
        self.decoder = nn.Sequential(nn.Linear(latent_dim, 256), nn.ReLU(), nn.Linear(256, 784), nn.Sigmoid())

    def forward(self, x):
        z_e = self.encoder(x)
        z_q, vq_loss, _ = self.vq(z_e)
        recon = self.decoder(z_q)
        return recon, vq_loss


vqvae = VQVAE().to(device)
opt_vq = optim.Adam(vqvae.parameters(), lr=1e-3)

vq_total, vq_recon, vq_vq = [], [], []
for epoch in range(12):
    vqvae.train()
    ep_t, ep_r, ep_v = 0, 0, 0
    for imgs, _ in train_loader:
        imgs = imgs.to(device)
        recon, vq_loss = vqvae(imgs)
        recon_loss = F.mse_loss(recon, imgs.view(-1, 784))
        loss = recon_loss + vq_loss
        opt_vq.zero_grad(); loss.backward(); opt_vq.step()
        ep_t += loss.item()
        ep_r += recon_loss.item()
        ep_v += vq_loss.item()
    n = len(train_loader)
    vq_total.append(ep_t/n)
    vq_recon.append(ep_r/n)
    vq_vq.append(ep_v/n)
    print(f"Epoch {epoch+1:02d} | Total: {ep_t/n:.4f} | Recon: {ep_r/n:.4f} | VQ: {ep_v/n:.4f}")

# Plotly VQ-VAE
fig = make_subplots(rows=1, cols=2, subplot_titles=("Losses", "VQ Loss"))
fig.add_trace(go.Scatter(y=vq_total, mode='lines+markers', name='Total', line=dict(color='indigo')), row=1, col=1)
fig.add_trace(go.Scatter(y=vq_recon, mode='lines+markers', name='Recon', line=dict(color='teal')), row=1, col=1)
fig.add_trace(go.Scatter(y=vq_vq, mode='lines+markers', name='VQ Loss', line=dict(color='darkorange')), row=1, col=2)
fig.update_layout(title="VQ-VAE Training Metrics", height=400, template="plotly_white")
fig.show()

print("✅ VQ-VAE تمام شد\n")


# ============================ CELL 4: VGAE ============================
print("="*70)
print("CELL 4: VGAE")
print("="*70)

!pip install torch_geometric -q

from torch_geometric.datasets import KarateClub
from torch_geometric.transforms import RandomLinkSplit
from torch_geometric.nn import GCNConv

dataset = KarateClub()
data = dataset[0]

transform_split = RandomLinkSplit(num_val=0.05, num_test=0.1, is_undirected=True, add_negative_train_samples=False)
train_data, val_data, test_data = transform_split(data)
train_data = train_data.to(device)

class GCNEncoder(nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels):
        super().__init__()
        self.conv1 = GCNConv(in_channels, hidden_channels)
        self.conv_mu = GCNConv(hidden_channels, out_channels)
        self.conv_logvar = GCNConv(hidden_channels, out_channels)

    def forward(self, x, edge_index):
        x = F.relu(self.conv1(x, edge_index))
        return self.conv_mu(x, edge_index), self.conv_logvar(x, edge_index)

class VGAE(nn.Module):
    def __init__(self, encoder):
        super().__init__()
        self.encoder = encoder

    def reparameterize(self, mu, logvar):
        if self.training:
            return mu + torch.randn_like(mu) * torch.exp(0.5 * logvar)
        return mu

    def encode(self, x, edge_index):
        mu, logvar = self.encoder(x, edge_index)
        z = self.reparameterize(mu, logvar)
        return z, mu, logvar

    def decode(self, z, edge_index):
        row, col = edge_index
        return torch.sigmoid((z[row] * z[col]).sum(dim=1))

vgae = VGAE(GCNEncoder(data.x.size(1), 32, 16)).to(device)
optimizer = optim.Adam(vgae.parameters(), lr=0.01)

vgae_losses, vgae_kls = [], []
for epoch in range(150):
    vgae.train()
    z, mu, logvar = vgae.encode(train_data.x, train_data.edge_index)
    # recon loss
    pos_loss = -torch.log(vgae.decode(z, train_data.edge_label_index) + 1e-8).mean()
    neg_edge = torch.randint(0, data.num_nodes, (train_data.edge_label_index.size(1), 2), device=device).t()
    neg_loss = -torch.log(1 - vgae.decode(z, neg_edge) + 1e-8).mean()
    recon_loss = pos_loss + neg_loss
    kl_loss = -0.5 * torch.mean(1 + logvar - mu.pow(2) - logvar.exp())
    loss = recon_loss + kl_loss

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    vgae_losses.append(loss.item())
    vgae_kls.append(kl_loss.item())

    if (epoch + 1) % 30 == 0:
        print(f"Epoch {epoch+1:03d} | Loss: {loss.item():.4f} | KL: {kl_loss.item():.4f}")

# Plotly VGAE
fig = make_subplots(rows=1, cols=2, subplot_titles=("Total Loss", "KL Divergence"))
fig.add_trace(go.Scatter(y=vgae_losses, mode='lines+markers', name='Total Loss', line=dict(color='teal')), row=1, col=1)
fig.add_trace(go.Scatter(y=vgae_kls, mode='lines+markers', name='KL', line=dict(color='darkorange')), row=1, col=2)
fig.update_layout(title="VGAE Training Progress", height=400, template="plotly_white")
fig.show()

print("\n🎉 تمام شد! همه ۴ نمودار Plotly آماده است.")
```
